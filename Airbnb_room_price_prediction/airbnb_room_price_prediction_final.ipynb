```python
data.drop(['id', 'name', 'host_id','neighbourhood'], axis=1, inplace =True)
```

```python
cat_col = data.select_dtypes(include=['object'])
cat_col
```

```python
num_col = data.select_dtypes(exclude=['object'])
num_col
```

```python
d1 = pd.get_dummies(cat_col,dtype=int,drop_first=True)
d1.head()
```

```python
d2 = pd.concat([num_col,d1], axis=1)
d2.head()
```

```python
x = d2.drop(['price'],axis=1)
y = d2['price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.preprocessing import RobustScaler
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
scaler = RobustScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

```python
#linear regression
```

```python
lin = LinearRegression()
lin.fit(x_train, y_train)
pr1 = lin.predict(x_test)
```

```python
diff1 = y_test-pr1
dt1 = pd.DataFrame({'Actual':y_test,'Predicted':pr1,'Difference':diff1})
dt1
```

```python
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,mean_squared_error,r2_score
print('mae :',mean_absolute_error(y_test,pr1))
print('mape :',mean_absolute_percentage_error(y_test,pr1))
sq = mean_squared_error(y_test,pr1)
print('mse :',sq)
print('rmse :',np.sqrt(sq))
print('r2score :',r2_score(y_test,pr1))
```

```python
#random forest
```

```python
random = RandomForestRegressor(n_estimators=100,criterion='poisson')
random.fit(x_train,y_train)
pr2 = random.predict(x_test)
```

```python
diff2 = y_test-pr2
dt2 = pd.DataFrame({'Actual':y_test,'Predicted':pr2,'Difference':diff2})
dt2
```

```python
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,mean_squared_error,r2_score
print('mae :',mean_absolute_error(y_test,pr2))
print('mape :',mean_absolute_percentage_error(y_test,pr2))
sq = mean_squared_error(y_test,pr2)
print('mse :',sq)
print('rmse :',np.sqrt(sq))
print('r2score :',r2_score(y_test,pr2))
```

```python
import xgboost as xg
```

```python
xgb_r = xg.XGBRegressor(objective ='reg:linear',n_estimators = 100, seed = 123)
xgb_r.fit(x_train,y_train)
pr3 = xgb_r.predict(x_test)
```

```python
diff3 = y_test-pr3
dt3 = pd.DataFrame({'Actual':y_test,'Predicted':pr3,'Difference':diff3})
dt3
```

```python
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,mean_squared_error,r2_score
print('mae :',mean_absolute_error(y_test,pr3))
print('mape :',mean_absolute_percentage_error(y_test,pr3))
sq = mean_squared_error(y_test,pr3)
print('mse :',sq)
print('rmse :',np.sqrt(sq))
print('r2score :',r2_score(y_test,pr3))
```

```python
search_space = {
    "n_estimators" : [100, 200, 500],
    "max_depth" : [3, 6, 9],
    "gamma" : [0.01, 0.1],
    "learning_rate" : [0.001, 0.01, 0.1, 1]
}
```

```python
GS = GridSearchCV(estimator = xgb_r,
                  param_grid = search_space,
                  scoring = ["r2", "neg_root_mean_squared_error"],
                  refit = "r2",
                  cv = 5,
                  verbose = 4)
```

```python
GS.fit(x_train, y_train)
```

```python
print(GS.best_params_)
```

```python
print(GS.best_score_)
```
