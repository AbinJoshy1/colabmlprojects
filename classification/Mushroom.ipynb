```python
ls1 = ['cap-shape', 'cap-surface', 'cap-color', 'bruises', 'odor',
       'gill-attachment', 'gill-spacing', 'gill-size', 'gill-color',
       'stalk-shape', 'stalk-root', 'stalk-surface-above-ring',
       'stalk-surface-below-ring', 'stalk-color-above-ring',
       'stalk-color-below-ring', 'veil-type', 'veil-color', 'ring-number',
       'ring-type', 'spore-print-color', 'population', 'habitat']
```

```python
from sklearn.preprocessing import LabelEncoder
lb  = LabelEncoder()
for i in ls1:
  d[i] = lb.fit_transform(d[i])
```

```python
d.head()
```

```python
y=d['class'].values
y
```

```python
x=d.iloc[:,1:].values
x
```

```python
from sklearn.model_selection import train_test_split
x_tr,x_te,y_tr,y_te = train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
s = StandardScaler()
s.fit(x_tr)
x_tr = s.transform(x_tr)
x_te = s.transform(x_te)
```

```python
from sklearn.neighbors import KNeighborsClassifier
k = KNeighborsClassifier(n_neighbors=7)
k.fit(x_tr,y_tr)
pr = k.predict(x_te)
```

```python
pr
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,ConfusionMatrixDisplay
conf = confusion_matrix(y_te,pr)
accu = accuracy_score(y_te,pr)
dis = ConfusionMatrixDisplay(conf)
```
