```python
import numpy as np
import pandas as pd
```

```python
outlook=['sunny','sunny','overcast','rainy','rainy','rainy',
         'overcast','sunny','sunny','rainy','sunny',
         'overcast','overcast','rainy']
temperature=['hot','hot','hot','mild','cool','cool','cool','mild',
             'cool','mild','mild','mild','hot','mild']
humidity=['high','high','high','high','normal','normal',
          'normal','high','normal','normal','normal','high','normal','high']

wind=['weak','strong','weak','weak','weak','strong','strong','weak',
      'weak','weak','strong','strong','weak','strong']

play=['no','no','yes','yes','yes','no','yes','no','yes','yes',
      'yes','yes','yes','no']
```

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
outlook_enc = lb.fit_transform(outlook)
temp_enc = lb.fit_transform(temperature)
hum_enc = lb.fit_transform(humidity)
wind_enc = lb.fit_transform(wind)
```

```python
fea = list(zip(outlook_enc,temp_enc,hum_enc,wind_enc))
fea
```

```python
from sklearn.naive_bayes import MultinomialNB
ml = MultinomialNB()
ml.fit(fea,play)
```

```python
pr = ml.predict([[2,0,0,0]])
pr
```
