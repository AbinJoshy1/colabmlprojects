```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
d = pd.read_csv('/content/diabetes.csv',sep=',')
d
```

```python
d.isna().sum()
```

```python
d.shape
```

```python
x = d.iloc[:,:-1].values
```

```python
y = d.iloc[:,-1].values
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30,random_state=1)
```

```python
x_train
```

```python
y_train
```

```python
y_test
```

```python
scalar=StandardScaler()
```

```python
scalar.fit(x_train)
```

```python
x_tr_norm = scalar.transform(x_train)
```

```python
x_te_norm = scalar.transform(x_test)
```

```python
x_tr_norm
```

```python
x_te_norm
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
k = KNeighborsClassifier(n_neighbors=9)
```

```python
k.fit(x_tr_norm,y_train)
```

```python
out = k.predict(x_te_norm)
```

```python
out
```

```python
y_test
```

performance evaluation

```python
from sklearn.metrics import confusion_matrix
m = confusion_matrix(y_test,out)
```

```python
m
```

```python
from sklearn.metrics import accuracy_score
sc = accuracy_score(y_test,out)
```
